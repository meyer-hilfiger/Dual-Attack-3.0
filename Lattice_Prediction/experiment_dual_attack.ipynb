```python
from sage.modules.misc import gram_schmidt

#Gaussian Heuristique
def GH(n):
    return float( sqrt((n/(2*pi*e)) * (pi*n)^(1/n)) )
```

```python
fd = open('out_n40_fft16_enum29.txt', 'r')
lines = fd.readlines()

tag = 0
N,n,q,kfft,enum = 0,0,0,0,0

Bprime_list = []
UinvT_list = []
dual_vectors = []
dual_vectors_rebased = []

for line in lines:
    if line[:3] == "-->":
        if tag == 2:
            line = line[:-1].replace("  ", " ")
            while "  " in line:
                line = line.replace("  ", " ")
            splitted_line = line.split(" ")
            n = int(splitted_line[3][2:])
            N = int(splitted_line[4][2:])
            q = int(splitted_line[5][2:])
            kfft = int(splitted_line[6][4:])
            enum = int(splitted_line[7][5:])

            print("q =",q)
            print("n =",n)
            print("N =",N)
            print("kfft =",kfft)
            print("enum =",enum)
        tag += 1
    elif(tag == 1):
        splitted_line = line[1:-2].split(", ")
        v = [float(x) for x in splitted_line]
        Bprime_list.append(v)
    elif(tag == 2):
        splitted_line = line[1:-2].split(", ")
        v = [float(x) for x in splitted_line]
        UinvT_list.append(v)
    elif(tag == 3):
        splitted_line = line[1:-2].split(", ")
        v = [float(x) for x in splitted_line]
        dual_vectors.append(v)

Bprime = matrix(Bprime_list)
Bprime_inv = Bprime.inverse()
UinvT = matrix(UinvT_list)

vol_primal = abs(Bprime.determinant())
vol_dual = abs(Bprime_inv.determinant())
print("Vol(dual) =",float(log(vol_dual, 2)))
print("Vol(primal) =",float(log(vol_primal, 2)))
print("estimation det(B') = ", float(log(q^(n/2) * 2^kfft, 2)))
print("estimation 1/det(B') = ", float(log(q^(-n/2) * 2^(-kfft), 2)))
dgv = float(GH(n)*(vol_primal^(1/n)))
print("dgv =",dgv)

lengths = [sqrt(sum([x*x for x in vector(v)*Bprime_inv.transpose()])) for v in dual_vectors]
w = float(sum(lengths)/len(lengths))
sigma_w = float(sqrt(sum([(x - w)^2 for x in lengths])/len(lengths)))
print("w =",w)
print("standard deviation of w =",sigma_w)
w_ = float(sqrt(4/3) * GH(n) * q^(-1/2) * 2^(-kfft/n))
print("w theorique : sqrt(4/3) * GH(n) * q^(-1/2) * 2^(-kfft/n) =",w_)

# histogramme distribution w
scale = 10000
max_w = max(lengths)
min_w = min(lengths)
distrib_w = [0]*scale
for w in lengths:
    distrib_w[round(w*scale)] += 1
print("\nDistribution w in W (precision = 10^(-4)):")
for x in range(round(min_w*scale), round(max_w*scale) + 1):
    print([float(x/scale), distrib_w[x]])
list_plot([[float(x/scale), distrib_w[x]] for x in range(round(min_w*scale), round(max_w*scale) + 1)], color="teal").show()


def Nj_f(x):
    card_Lambda = q^(n/2) * 2^(kfft)
    return float((1/card_Lambda) * (x/GH(n))^n)

#print("\nNj in function of j:")
#plot(Nj_f, 0, dgv).show()

L = []
L_pt = []


j = dgv - 70.0
while Nj_f(j) < 1.0 :
    j += 1
    
    L_tmp = []
    x = 0.0
    for i in range(round(min_w*scale), round(max_w*scale) + 1):
        w = float(i/scale)
        N = distrib_w[i]
        bessel = bessel_J(n/2 - 1, 2*pi*w*j)
        cst = float(N*sqrt(pi * n)/e)
        T = float((n/(2*pi*e*w*j))^(n/2 - 1))
        
        L_tmp.append([w, float(cst*T*bessel)])
        x += float(cst*T*bessel)
    
    #print("\n Nb(w_i)*sqrt(pi * n)/e * (n/(2*pi*e*w_i*j))^(n/2 - 1) * bessel_J(n/2 - 1, 2*pi*w_i*j) in function of w_i:")
    #list_plot(L_tmp, legend_label="j = "+str(j)).show()
    Nj = Nj_f(j)
    y = float(log(Nj, 2))
    
    L.append([x,y])

#(plot(lambda x: bessel_J(n/2 - 1, 2*pi*w*x), 0, dgv, color='green') + list_plot(L_pt, color='red')).show()

plt = list_plot(L, plotjoined=True,  linestyle='-.', color='red', axes_labels=['$x$', '$\\mathbb{P}\\left(f_{\\mathcal{W}}(\\mathbf{y}) \\geq x \\right)$'], legend_label='this paper', ticks=[[20,40,60,80,100], [-10,-20,-30,-40]], tick_formatter=[["$20$","$40$","$60$","$80$","$100$"], ["$2^{-10}$","$2^{-20}$","$2^{-30}$","$2^{-40}$"]], fontsize=12)
plt += plt_DP23_n40
plt.set_legend_options(handlelength=2, loc=(0.5,0.7), font_size=12)
plt.show()
#plt.save("figures/dual_attack_n40.pdf")
```
